```python
def create_NMM(G, 
               He = 3.25, # Synaptic gain in mV
               Hi = 22, # Synaptic gain in mV
               Te = 10, # excitatory time constant
               Ti = 20, # inhibitory time constant 
               c11 = 135, # Average number of contacts, c = 135
               c12 = .8 * 135, # Average number of contacts, c = 135
               c21 = .25 * 135, # Average number of contacts, c = 135
               c22 = .25 * 135, # Average number of contacts, c = 135
               c31 = 1, # Average number of contacts
               c32 = 1, # Average number of contacts
               v0 = 6, # mV
               e0 = 5, # s^-1
               r = .56, # mV ^-1
               v1 = 1, # state
               v2 = 1, # state
               v3 = 1, # state
               x1 = 1, # state
               x2 = 1, # state
               x3 = 1): #state
    
    # sigmoid function for NMM
    def S(x, c1, c2, v0 , e0, r):
        num = c1 * e0
        den = 1 + np.exp(r*(v0 - c2*x))
        return num / den
    
    n = len(G)  # number of nodes
    # Initialization of all parameters for each node, look at netwrok_modules for create_vec_states function description
    parameters = nm.create_vec_states([(n, He, .1 * He, 'normal'), (n, Hi, .1 * Hi, 'normal'), (n, Te, .1 * Te, 'normal'),
                                       (n, Ti, .1 * Ti, 'normal'), (n, c11, .1 * c11, 'normal'), (n, c12, .1 * c12, 'normal'),
                                       (n, c21, .1 * c21, 'normal'), (n, c22, .1 * c22, 'normal'), (n, c32, .1 * c32, 'normal'),
                                       (n, v0, .1 * v0, 'normal'), (n, e0, .1 * e0, 'normal'), (n, r, .1 * r, 'normal')])
    # Initial states
    states = nm.create_vec_states([(n, v1, .1 * v1, 'normal'), (n, v2, .1 * v2, 'normal'), (n, v3, .1 * v3, 'normal'),
                                   (n, x1, .1 * v1, 'normal'), (n, x2, .1 * x2, 'normal'), (n, x3, .1 * x3, 'normal')])
    
    # curry node function to be unique to each node: each node has a unique intrinsic frequency and amplitude
    def curry_node(He, Hi, Te, Ti, c11, c12, c22, c31, c32, v0, e0, r):
        def node(x):   # node function
            x0 = He/Te*(S(x[4], c11, c12, v0, e0, r)) - 2 * x[0] / Te - x[3] / (Te **2)
            x1 = Hi/Ti*(S(x[5], c21, c22, v0, e0, r)) - 2 * x[1] / Ti - x[4] / (Ti **2)
            x2 = He/Te*(S((x[3] - x[4]), c31, c32, v0, e0, r)) - 2 * x[2] / Te - x[5] / (Te **2)
            x3 = x[0]
            x4 = x[1]
            x5 = x[2]
            return np.array([x0, x1, x2, x3, x4, x5])
        return node

    
    h = [curry_node(He, Hi, Te, Ti, c11, c12, 
                    c22, c31, c32, v0, e0, r) 
         for He, Hi, Te, Ti, c11, c12, c22, c31, c32, v0, e0, r in zip(parameters[:, 0], parameters[:, 1], parameters[:, 2], 
                                      parameters[:, 3], parameters[:, 4], parameters[:, 5], 
                                      parameters[:, 6], parameters[:, 7], parameters[:, 8], 
                                      parameters[:, 9], parameters[:, 10], parameters[:, 11])]

    # curry coupling function for each node
    def curry_couple():
        def couple(x, y):  # coupling function
            return y-x
        return couple
    f = [curry_couple() for K in parameters[:, 2]]

    NMM_model = nm.nmodel(G, states, h, f, M, N, dt=.01)
    return NMM_model

NMM = create_NMM(G)
NMM.run(.01)
nm.state_course(NMM.x[:,0,:])
print(NMM.x[:,:,-1].shape)
```
